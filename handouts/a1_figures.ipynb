```python
%matplotlib inline
import numpy as np
import scipy as sp
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import IPython as ip
mpl.style.use('ggplot')
# mpl.rc('font', family='Noto Sans CJK TC')
ip.display.set_matplotlib_formats('svg')
```

**NOTE:** The FDR and FOR below are *predefined* FDR and FOR, actually.

```python
alpha_v = np.arange(0.01, 0.1+0.01, 0.01)
beta = 0.2
effect_size_v = np.arange(0.1, 0.5+0.05, 0.05)

sample_size_m = np.zeros((len(alpha_v), len(effect_size_v)))
for i in range(len(alpha_v)):
    for j in range(len(effect_size_v)):
        sample_size_m[i, j] = sm.stats.tt_ind_solve_power(
            alpha=alpha_v[i],
            power=1-beta,
            effect_size=effect_size_v[j],
            ratio=1,
            nobs1=None,
        )

plt.figure(figsize=(6*1.2, 4*1.2))
sns.heatmap(
    sample_size_m,
    vmax=300, vmin=0, cmap='Greens', annot=True, fmt='.0f',
    yticklabels=[f'{a:.2f}' for a in alpha_v],
    xticklabels=[f'{e:.2f}' for e in effect_size_v],
    cbar_kws=dict(label='Sample Size'),
)
plt.ylabel(f'$ \\alpha,\,\\beta = {beta:.2f} $')
plt.xlabel('Effect Size')
plt.savefig('figures/a1_alpha_effect_size_sample_size.png', bbox_inches='tight', dpi=600)
```

```python
alpha_v = np.arange(0.001, 0.01+0.001, 0.001)
beta = 0.01
effect_size_v = np.arange(0.01, 0.1+0.01, 0.01)

sample_size_m = np.zeros((len(alpha_v), len(effect_size_v)))
for i in range(len(alpha_v)):
    for j in range(len(effect_size_v)):
        sample_size_m[i, j] = sm.stats.tt_ind_solve_power(
            alpha=alpha_v[i],
            power=1-beta,
            effect_size=effect_size_v[j],
            ratio=1,
            nobs1=None,
        )

plt.figure(figsize=(6*1.2, 4*1.2))
sns.heatmap(
    sample_size_m,
    vmax=50000, vmin=0, cmap='Greens', annot=True, fmt='.0f',
    yticklabels=[f'{a:.3f}'[1:] for a in alpha_v],
    xticklabels=[f'{e:.2f}' for e in effect_size_v],
    cbar=False,
    # cbar_kws=dict(label='Sample Size'),
)
plt.ylabel(f'$ \\alpha,\,\\beta = {beta:.2f} $')
plt.xlabel('Effect Size')
plt.savefig('figures/a1_alpha_effect_size_sample_size_2.png', bbox_inches='tight', dpi=600)
```

```python
# f: alpha, beta=0.20, anr -> fdr

alpha_v = np.arange(0.1, 0, -0.01)
beta = 0.20
anr_v = np.arange(1, 0.5-0.05, -0.05)

power = 1-beta
apr_v = 1-anr_v

fdr_m = alpha_v*anr_v[:, None] / (alpha_v*anr_v[:, None] + power*apr_v[:, None])

plt.figure(figsize=(6*1.2, 4*1.2))
sns.heatmap(
    fdr_m,
    vmin=0, vmax=1, cmap='Reds', annot=True, fmt='.2f',
    yticklabels=[f'{p:.2f}' for p in anr_v],
    xticklabels=[f'{a:.2f}' for a in alpha_v],
    cbar_kws=dict(label='False Discovery Rate'),
)
plt.ylabel('Actaul Negative Rate')
plt.xlabel(f'$ \\alpha,\,\\beta = {beta:.2f} $')
plt.savefig('figures/a1_anr_alpha_fdr.png', bbox_inches='tight', dpi=600)
```

```python
# f: alpha, beta=0.01, anr -> fdr

alpha_v = np.arange(0.01, 0, -0.001)
beta = 0.01
anr_v = np.arange(1, 0.9-0.01, -0.01)

power = 1-beta
apr_v = 1-anr_v

fdr_m = alpha_v*anr_v[:, None] / (alpha_v*anr_v[:, None] + power*apr_v[:, None])

plt.figure(figsize=(6*1.2, 4*1.2))
sns.heatmap(
    fdr_m,
    vmin=0, vmax=1, cmap='Reds', annot=True, fmt='.2f',
    yticklabels=[f'{p:.2f}' for p in anr_v],
    xticklabels=[f'{a:.3f}' for a in alpha_v],
    cbar_kws=dict(label='False Discovery Rate'),
)
plt.ylabel('Actaul Negative Rate')
plt.xlabel(f'$ \\alpha,\,\\beta = {beta:.2f} $')
plt.savefig('figures/a1_anr_alpha_fdr_2.png', bbox_inches='tight', dpi=600)
```

```python
# f: alpha=0.05, beta, anr -> for

alpha = 0.05
beta_v = np.arange(0.2, 0, -0.02)
anr_v = np.arange(0.5, 0-0.05, -0.05)

cl = 1-alpha
apr_v = 1-anr_v

for_m = beta_v*apr_v[:, None] / (cl*anr_v[:, None] + beta_v*apr_v[:, None])

plt.figure(figsize=(6*1.2, 4*1.2))
sns.heatmap(
    for_m,
    vmin=0, vmax=1, cmap='Blues', annot=True, fmt='.2f',
    yticklabels=[f'{p:.2f}' for p in anr_v],
    xticklabels=[f'{b:.2f}' for b in beta_v],
    cbar_kws=dict(label='False Omission Rate'),
)
plt.ylabel('Actaul Negative Rate')
plt.xlabel(f'$ \\alpha = {alpha:.2f},\,\\beta $')
plt.savefig('figures/a1_anr_beta_for.png', bbox_inches='tight', dpi=600)
```
